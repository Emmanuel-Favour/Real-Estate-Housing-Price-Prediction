#### Considering PolynomialFeature

```python
from sklearn.preprocessing import PolynomialFeatures
rsqr = []
degree = [1,2,3,4]
for n in degree:
    pr = PolynomialFeatures(degree=n)
    x_train_pr = pr.fit_transform(x_train)
    x_text_pr = pr.fit_transform(x_test)
    poly = LinearRegression()
    poly.fit(x_train_pr, y_train)
    poly.fit(x_train_pr, y_train)
    rsqr.append(poly.score(x_text_pr, y_test))

plt.plot(degree, rsqr)
plt.xlabel('order')
plt.ylabel('R^2')
plt.title('R^2 Using Test Data')
plt.text(3, 0.75, 'Maximum R^2 ')
```

From the chart above, it can be observed that order 3.0 has the best R^2 Score, but this has to be further validated

Cross validation helps to find out how the model performs across folds to detect which of the polynomial degrees generalizes well

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
for d in [1,2,3,4]:
    model = Pipeline([("Poly", PolynomialFeatures(degree =d)), ("linreg", LinearRegression())])
    score = cross_val_score(model, x_train, y_train, cv=3, scoring="r2")
    print(score)
    print(f"Degree {d}:Mean={score.mean():.3f}, std={score.std():.3f}" )
```

From the cross validation, it can be seen that degree 2 maintains a higher mean, and the R^2 are consistent across folds. Despite the fact that degree 3 had an R^2 score of 81% but can not generalize well makes it a NO

```python
from tqdm import tqdm
rsqr_train =[]
rsqr_test = []
dummy1 = []
Alpha = np.array(range(0,1000))
pbar= tqdm(Alpha)
for alpha in pbar:
    RidgeModel= Ridge(alpha=alpha)
    RidgeModel.fit(x_train_pr, y_train)
    train_score, test_score = RidgeModel.score(x_train_pr, y_train), RidgeModel.score(x_test_pr, y_test)
    
    pbar.set_postfix({"TrainScore":train_score, "TestScore":test_score})

    rsqr_train.append(train_score)
    rsqr_test.append(test_score)
```

```python
plt.figure(figsize=(15,12))

plt.plot(Alpha, rsqr_train, label="Model Training")
plt.plot(Alpha, rsqr_test, label= "Model Testing")
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.legend()
```

```python
# Define parameter grid (notice 'ridge__alpha' for Ridge inside pipeline)
param_grid = {
    "ridge__alpha": [0.001, 0.1, 1, 10, 100, 1000, 10000, 100000]
}
```

```python
# Define pipeline
pipeline = Pipeline([
    ("poly", PolynomialFeatures(degree=3)),  # polynomial feature           
    ("ridge", Ridge())                       # ridge regression (alpha tuned later)
])
```

```python
# GridSearchCV
grid = GridSearchCV(pipeline, param_grid, cv=4, scoring="r2")
grid.fit(x_train, y_train)
```

```python
# Best model
best_rr = grid.best_estimator_
print("Best Parameters:", grid.best_params_)
print("Best CV Score:", grid.best_score_)
```

```python
# Test score
test_score = best_rr.score(x_test, y_test)
print("Test R2 Score:", test_score)
```

Initially, polyFeature degree 2 makes a strong case as a Baseline Model as it generalizes better than the degree 3 despite the high R^2 of degree 3, but upon regularization with Ridge regression degree 3 also now seem to have a good CV score 

### Serialization

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import Ridge
#Final Model Pipeline
best_model = Pipeline([
    ("scaler", StandardScaler()),
    ("poly", PolynomialFeatures(degree=3)),
    ("ridge", Ridge(alpha=1000))
])

best_model.fit(x_train, y_train)

print("Train R2:", best_model.score(x_train, y_train))
print("Test R2:", best_model.score(x_test, y_test))
```

```python
from fastapi import FastAPI
from pydantic import BaseModel
import joblib
import numpy as np

#Load model
model = joblib.load("ridge_poly3_model.pk1")

#Initialize API
app = FastAPI(tittle="Real Estate Price Prediction", version="1.0")

#Define Request Schema
class InputData(BaseModel):
    features: list

@app.post("/predict")
def predict(data: InputData):
    # Convert to numpy array and reshape
    X = np.array(data.features).reshape(1, -1)
    
    # Predict
    prediction = model.predict(X)
    
    return {"prediction": prediction.tolist()}
```
